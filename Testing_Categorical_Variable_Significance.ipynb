```python
# Importing required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import statsmodels.api as sm
```

```python
# Creating a dataframe
transport_df = pd.read_csv("Transportation_Hash_Dummy_Encoded_Dataset.csv")
transport_df.tail()
```

```python
transport_df.info()
```

```python
transport_df.columns
```

```python
# Step 1: Fit the Full Model
# Assuming df is your DataFrame with columns 'numerical_var1', 'numerical_var2', 'C1', 'C2', and 'target_variable'
X_full = transport_df[['make', 'model', 'engine_size', 'cylinders', 'fuel_consumption_comb', 'vehicle_class_cluster1', 'vehicle_class_cluster2', 'vehicle_class_cluster3', 'vehicle_class_cluster4', 'vehicle_class_cluster5', 'vehicle_class_cluster6', 'transmission_AM', 'transmission_AS', 'transmission_AV', 'transmission_M', 'fuel_type_E', 'fuel_type_X', 'fuel_type_Z']]
X_full = sm.add_constant(X_full)  # Add constant term for intercept
y = transport_df['co2_emissions']
full_model = sm.OLS(y, X_full).fit()
full_model.summary()
```

```python
# , 'model', 'cylinders', 'engine_size', 'fuel_consumption_comb', 'vehicle_class_cluster1', 'vehicle_class_cluster2', 'vehicle_class_cluster3', 'vehicle_class_cluster4', 'vehicle_class_cluster5', 'vehicle_class_cluster6', 'transmission_AM', 'transmission_AS', 'transmission_AV', 'transmission_M', 'fuel_type_E', 'fuel_type_X', 'fuel_type_Z'
# Step 2: Fit the Reduced Model
X_reduced = transport_df[['make']]
X_reduced = sm.add_constant(X_reduced)  # Add constant term for intercept
reduced_model = sm.OLS(y, X_reduced).fit()
reduced_model.summary()
```

```python
# Step 3: Perform the Test
# For ANOVA
anova_results = sm.stats.anova_lm(reduced_model, full_model)
f_statistic = anova_results['F'][1]  # Extract F-statistic
p_value_anova = anova_results['Pr(>F)'][1]  # Extract p-value
anova_results
```

```python
# Step 4: Interpret the Results
alpha = 0.05
if p_value_anova < alpha:
    print("ANOVA: Reject the null hypothesis. Categorical variables are jointly significant.")
else:
    print("ANOVA: Fail to reject the null hypothesis. Categorical variables are not jointly significant.")
```
